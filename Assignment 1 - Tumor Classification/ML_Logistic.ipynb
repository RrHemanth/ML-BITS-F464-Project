```python
def train_test_split(data, test_ratio=0.33):

    
    # split the data into training and testing sets
    split_idx = int(len(data) * (1 - test_ratio))
    train_data = data[:split_idx]
    test_data = data[split_idx:]
    
    return train_data, test_data
```

# Log reg code

```python
class LogisticRegression:
    def __init__(self, n_iterations, learning_rate):
        self.n_iterations = n_iterations
        self.learning_rate = learning_rate
        self.parameters = {}
        
    def sigmoid(self, x):
        s = 1/(1 + np.nan_to_num(np.exp(-x)))
        return s

    def initialize_parameters(self, n_features):
        W = np.zeros((1, n_features))
        b = np.zeros((1, 1))

        self.parameters = {
            'W': W,
            'b': b
        }

    def forward_propagation(self, X):
        W = self.parameters['W']
        b = self.parameters['b']

        Z = np.dot(W, X) + b
        A = self.sigmoid(Z)

        results = {
            'Z': Z,
            'A': A
        }

        return results

    def compute_cost(self, A, Y):
        m = Y.shape[1]
        
        A = np.nan_to_num(A)
        logA = np.nan_to_num(np.log(A))
        logA_ = np.nan_to_num(np.log(1-A))
        
        cost = (-1/m)*np.nan_to_num(np.dot(Y, logA.T) + np.dot(1-Y, logA_.T))
        cost = np.squeeze(cost)
    
        return cost

    def backward_propagation(self, A, X, Y):
        m = Y.shape[1]
        dZ = A - Y
        dW = np.dot(dZ, X.T)/m
        db = np.sum(dZ, axis=1, keepdims=True)

        results = {
            "dZ": dZ,
            "dW": dW,
            "db": db
        }

        return results

    def gradient_descent(self, X, Y):
        import matplotlib.pyplot as plt    
        errors=[]
        for i in range(self.n_iterations):
            A = self.forward_propagation(X)['A']
            if (i+1)%100 == 0:
                print(f"Cost after {i+1} iterations: {self.compute_cost(A, Y)}")
            errors.append(self.compute_cost(A, Y))
            
            gradients = self.backward_propagation(A, X, Y)
            self.parameters['W'] = self.parameters['W'] - self.learning_rate*gradients['dW']
            self.parameters['b'] = self.parameters['b'] - self.learning_rate*gradients['db']
        plt.plot(errors)
        plt.ylabel('cost')
        plt.xlabel('iterations (per hundreds)')
        plt.title("Learning rate =" + str(self.learning_rate) )
        plt.show()        
        
    def fit(self, X, Y):
        
        X = np.array(X).T
        Y = np.array(Y).reshape(1, -1)

        n_features = X.shape[0]
        self.initialize_parameters(n_features)

        self.gradient_descent(X, Y)

    def predict(self, X):
        X = np.array(X).T
        m = X.shape[1]
        
        W = self.parameters['W']
        b = self.parameters['b']

        Z = np.dot(W, X) + b
        A = self.sigmoid(Z)

        predictions = []
        for i in range(m):
            if A[0, i] > 0.7:  #threshold
                predictions.append(1)
            else:
                predictions.append(0)

        return predictions
```

```python
class LogisticRegression_minibatch:
    def __init__(self, learning_rate=0.01, num_iterations=1000, batch_size=32):
        self.learning_rate = learning_rate
        self.num_iterations = num_iterations
        self.batch_size = batch_size
        self.weights = None
        self.bias = None
        
    def sigmoid(self, z):
        """Sigmoid activation function."""
        return 1 / (1 + np.exp(-z))
    
    def initialize_parameters(self, n_features):
        """Initialize weights and bias."""
        self.weights = np.zeros((n_features, 1))
        self.bias = 0
        
    def forward_propagation(self, X):
        """Compute forward propagation."""
        Z = np.dot(X, self.weights) + self.bias
        A = self.sigmoid(Z)
        return A
    
    def compute_cost(self, A, Y):
        """Compute cost function."""
        m = Y.shape[0]
        cost = (-1/m) * np.sum(Y*np.log(A) + (1-Y)*np.log(1-A))
        return cost
    
    def backward_propagation(self, X, A, Y):
        """Compute gradients using backward propagation."""
        m = Y.shape[0]
        dZ = A - Y
        dW = (1/m) * np.dot(X.T, dZ)
        db = (1/m) * np.sum(dZ)
        return dW, db
    
    def update_parameters(self, dW, db):
        """Update weights and bias."""
        self.weights = self.weights - self.learning_rate * dW
        self.bias = self.bias - self.learning_rate * db
    
    def fit(self, X, Y):
        """Train the model using mini-batch gradient descent."""
        n_features = X.shape[1]
        self.initialize_parameters(n_features)
        costs = []
        m = X.shape[0]
        for i in range(self.num_iterations):
            # Shuffle the data
            permutation = np.random.permutation(m)
            shuffled_X = X[permutation]
            shuffled_Y = Y[permutation]
            # Split the data into mini-batches
            for j in range(0, m, self.batch_size):
                batch_X = shuffled_X[j:j+self.batch_size]
                batch_Y = shuffled_Y[j:j+self.batch_size]
                A = self.forward_propagation(batch_X)
                cost = self.compute_cost(A, batch_Y)
                dW, db = self.backward_propagation(batch_X, A, batch_Y)
                self.update_parameters(dW, db)
            if i % 100 == 0:
                print("Cost after iteration {}: {}".format(i, cost))
            costs.append(cost)
        
        plt.plot(costs)
        plt.ylabel('cost')
        plt.xlabel('iterations (per hundreds)')
        plt.title("Learning rate =" + str(self.learning_rate) )
        plt.show()
        return costs
    
    def predict(self, X):
        """Predict class labels for input data."""
        A = self.forward_propagation(X)
        Y_prediction = np.zeros((X.shape[0], 1))
        Y_prediction[A > 0.7] = 1
        return Y_prediction
```
