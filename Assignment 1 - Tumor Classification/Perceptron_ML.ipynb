```python
y=df[['diagnosis']]
X=df.loc[0:563,'radius_mean':]
X=X.sample(frac=1, axis=1).reset_index(drop=True)
X_train = X.loc[0:377].values
X_test = X.loc[378:563].values
y_train = y.loc[0:377]
y_train = y_train['diagnosis'].tolist()
y_test = y.loc[378:563]
y_test = y_test['diagnosis'].tolist()
X
```

```python
w=[0]*30
print(w)
epoch = 10500
```

```python
for j in range(epoch):
    for i in range(len(y_train)):
        if (y_train[i]*(X_train[i].dot(w))<=0):
            w=w+y_train[i]*X_train[i]
        elif (y_train[i]*(X_train[i].dot(w))>0):
            w=w+0
print(w)        
```

```python
y_pred=[]
for i in range(len(y_test)):
    if((X_test[i].dot(w))>0):
        y_pred.append(1)
    elif((X_test[i].dot(w))<0):
        y_pred.append(-1)
         
print(y_pred)          
```

```python
count=0
t=0
recall = 0
precision = 0
for i in range(len(y_test)):
    if (y_test[i]==y_pred[i]==1):
        count=count+1
        t=t+1
    elif(y_test[i]==y_pred[i]==-1):
        count=count+1
    elif(y_pred[i] == 1):
        recall = recall +1
    elif(y_pred[i] == -1):
        precision = precision +1

recall = recall + t
precision = precision +t
print('Accuracy',count/len(y_test))   
print('Recall',t/recall)
print('Precision',t/precision)
```
