```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('adult.csv')
df
```

```python
df.columns
```

```python
df[' <=50K'].value_counts()
```

```python
df[' <=50K'].replace({' <=50K':0, ' >50K':1}, inplace = True)
df
```

```python
df[' <=50K'].value_counts()
```

```python
df.rename(columns={' <=50K':'Income'}, inplace=True)
df
```

```python
df.isnull().sum()
```

```python
df = df.sample(frac=1, random_state=42)
```

```python
int_columns = df.select_dtypes(include=["int"]).columns
```

```python
df[int_columns] = (df[int_columns] - df[int_columns].min()) / (df[int_columns].max() - df[int_columns].min())
```

```python
X = df.drop("Income", axis=1)
y = df["Income"]
```

```python
train_size = int(0.67 * len(df))
X_train, X_test = X[:train_size], X[train_size:]
y_train, y_test = y[:train_size], y[train_size:]
```

```python
def compute_prior_probabilities(y):
    classes, counts = np.unique(y, return_counts=True)
    total_count = len(y)
    prior_probs = {}
    for i in range(len(classes)):
        prior_probs[classes[i]] = (counts[i] + 1) / (total_count + len(classes))
    return prior_probs
```

```python
def compute_likelihood_probabilities(X, y):
    classes = np.unique(y)
    feature_probs = {}
    for c in classes:
        c_X = X[y == c]
        feature_probs[c] = {}
        for col in c_X.columns:
            col_counts = c_X[col].value_counts()
            col_prob = {}
            for index, count in col_counts.items():
                col_prob[index] = ((c_X.sum(axis=0) + 1) / (len(c_X) + 2)
            feature_probs[c][col] = col_prob
    return feature_probs
```

```python
def predict(X, prior_probs, feature_probs):
    predictions = []
    for i in range(len(X)):
        class_probs = {}
        for c in prior_probs.keys():
            class_probs[c] = prior_probs[c]
            for col in X.columns:
                class_probs[c] *= feature_probs[c][col].get(X.iloc[i][col], 1 / len(feature_probs[c][col]))
        predictions.append(max(class_probs, key=class_probs.get))
    return predictions
```

```python
prior_probs = compute_prior_probabilities(y_train)
```

```python
feature_probs = compute_likelihood_probabilities(X_train, y_train)
```

```python
predictions = predict(X_test, prior_probs, feature_probs)
```

```python
accuracy = np.sum(predictions == y_test) / len(y_test)
print(accuracy)
```

```python
from sklearn.metrics import classification_report

# Make predictions on test set
predictions = predict(X_test, prior_probs, feature_probs)

# Compute classification report
report = classification_report(y_test, predictions)

print(report)
```
